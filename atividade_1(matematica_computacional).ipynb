```python
def plotagem(matA,matB,titulo):
  plt.arrow(0, 0, matA[0][0], matA[0][1], color='blue',linewidth=1,head_width=0.08,head_length=0.1,length_includes_head=True,label='Vetor A'), plt.arrow(0, 0, matA[1][0], matA[1][1], color='red',linewidth=1,head_width=0.08,head_length=0.1,length_includes_head=True,label='Vetor B')
  plt.title(f'{titulo}'), plt.axis('equal'), plt.legend()
  novo_grid = deformar_grid(matB)
  for linha in novo_grid:
   plt.plot(linha[0, :], linha[1, :], color='gray', alpha=0.6, linewidth=1)
  plt.show()
def plotagem_sem_multi(vetA,vetB,titulo):
  plt.arrow(0, 0, vetA[0], vetA[1], color='blue',linewidth=1,head_width=0.08,head_length=0.1,length_includes_head=True,label='Vetor A'), plt.arrow(0, 0, vetB[0], vetB[1], color='red',linewidth=1,head_width=0.08,head_length=0.1,length_includes_head=True,label='Vetor B')
  plt.title(f'{titulo}')
  plt.axis('equal')
  plt.grid(),plt.legend()
  plt.show()
def deformar_grid(matB):
  limite,intervalo,lista_de_grids_deformados = 5,1,[]
  vetores_espaco = np.arange(-limite, limite + 1, intervalo)
  for i in vetores_espaco:
    linha_y,  linha_x = np.array([vetores_espaco, [i] * len(vetores_espaco)]), np.array([[i] * len(vetores_espaco), vetores_espaco])
    linha_deformando_y, linha_deformando_x = matB @ linha_y, matB @ linha_x
    lista_de_grids_deformados.append(linha_deformando_y), lista_de_grids_deformados.append(linha_deformando_x)
  return lista_de_grids_deformados
def multiplica_matriz(matA,matB,titulo):
  plotagem(matA@matB,matB,titulo)
  print("\t \n o determinante da matriz é:\n", np.linalg.det(matA@matB))
  return (matA@matB)
def soma_e_multi_matrizes(matA,matB,titulo):
  matA[:, 0] *= np.sum(matB[0,:])
  try:
    matA[:, 1] *= np.sum(matB[1,:])
  except:
    pass
  rows_sumed = np.sum(matA, axis=1)
  plotagem_sem_multi(rows_sumed,[0,0],titulo)

def noventa_graus_rot(vetA,titulo):
  matriz_de_rot = np.array([[0, 1],
                                [-1,  0]])
  vetor_1d = (matriz_de_rot@vetA).tolist()
  plotagem_sem_multi([6,-1],[0,0],titulo)
  return (matriz_de_rot@vetA)



```

```python
#1
mat1,mat2,mat2_1,mat2_2,x  = np.array([[0,2],[1,0]]), np.array([[1,-2],[1,0]]), np.array([[1],[1]]), np.array([[-2],[0]]),np.array([[1],[6]])
plotagem_sem_multi(mat1[0],mat1[1],"#1 matriz rotacionada em 90º")
multiplica_matriz(mat1,mat2,"#1 matriz rotacionada em 90º com shear")
#2
# o determinante é negativo pois a matriz está espalhada, pois foi rotacionada em 90 graus
#3
soma_e_multi_matrizes(mat1.copy(),mat2_1,"#3 achando o I da matriz")
soma_e_multi_matrizes(mat1.copy(),mat2_2,"#3 achando o J da matriz")
#4
multiplica_matriz(mat2,mat1,"#4 matriz rotacionada em 90º com shear")
soma_e_multi_matrizes(mat2_1.copy(),mat1,"#4 achando o I da matriz")
soma_e_multi_matrizes(mat2_2.copy(),mat1,"#4 achando o J da matriz")

#5
plotagem_sem_multi(x.flatten(),[0,0],"vetor x em estado inicial")
y = noventa_graus_rot(x,"vetor x rotacionado em 90º")
print("produto escalar x e y: \t",x.flatten()@y.flatten())

```

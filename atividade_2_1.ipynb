### Métodos de Isolamento de Raízes

* **Bisseção (`bissecao`)**: É um método de busca intervalar que divide o intervalo $[a, b]$ ao meio sucessivamente. Ele requer que $f(a) \cdot f(b) < 0$ (Teorema de Bolzano). É extremamente robusto e sempre converge, embora de forma lenta.

* **Falsa Posição (`fake_pos`)**: Semelhante à bisseção, mas em vez de usar o ponto médio, utiliza a média ponderada baseada nos valores de $f(a)$ e $f(b)$ (uma linha reta ligando os pontos). Geralmente converge mais rápido que a bisseção ao "puxar" a estimativa para onde a função está mais próxima de zero.

* **Newton (`newton`)**: Um método de refinamento que utiliza a derivada $f'(x)$ para encontrar a tangente à curva no ponto atual. Onde essa tangente cruza o eixo $x$ torna-se a próxima estimativa. É muito rápido (convergência quadrática), mas exige o cálculo da derivada e uma boa estimativa inicial.

* **Secante (`secante`)**: Uma variação do método de Newton que substitui a derivada pela inclinação da reta secante que passa por dois pontos anteriores ($x_0$ e $x_1$). É ideal para quando o cálculo da derivada da função é difícil ou impossível.

* **Ponto Fixo (`ponto_fixo`)**: Transforma a equação $f(x) = 0$ em uma forma equivalente $x = g(x)$. O método consiste em iterar $x_{i+1} = g(x_i)$ até que o valor se estabilize. A convergência depende da função de iteração escolhida satisfazer os critérios do Teorema do Ponto Fixo.

### Comparativo de Aplicação dos Métodos

* **Bisseção**: É o "seguro de vida". Sempre converge se houver troca de sinal, mas como você notou (chegando a 19 interações), é o mais lento. Use quando a estabilidade for mais importante que a velocidade.

* **Falsa Posição**: Tenta ser uma Bisseção inteligente. Geralmente reduz o número de interações (caiu de 19 para 4 em um dos seus testes) ao inclinar a busca para onde o valor de $f(x)$ é menor.

* **Newton**: É o "velocista". Quando funciona, resolve em 1 ou 2 iterações. Porém, exige que você conheça a derivada da função e que o chute inicial seja bom, caso contrário, ele pode divergir (falhar).

* **Secante**: É o Newton para quando você não tem a derivada. Quase tão rápido quanto ele, mas usa dois pontos anteriores para aproximar a inclinação da curva.

## Estrutura de Laços (Loops)

O algoritmo é construído sore duas camadas principais de repetição:

* **Laço Externo (Convergência):** Controlado pela variável `itmax`. Ele representa as tentativas globais do método ara encontrar a raiz. A cada iteração deste laço, o valor de `x` é refinado.
* **Laço Interno (Avaliação Polinomial):** Itera sobre os coeficientes do polinômio ($a_n, a_{n-1}, \dots, a_1$). Ele utiliza a estratégia de deltax para calcular simultaneamente:
    1.  O valor da função no ponto atual ($b$).
    2.  O valor da derivada no ponto atual ($c$).

## Adaptação para o Método do Ponto Fixo (ponto_fixo_delta)

### Como funciona:
O MPF não busca a raiz diretaente em $f(x)=0$, mas sim o ponto onde $x = \phi(x)$. Não há necessidade de calcular derivadas ou usar a estrutura de deltax para coeficientes.

### Alterações realizadas:
* **Remoção do Laço :** O laço `Para i = (n-1)...1` é removido, pois não estamos mais tratando espcificamente polinômios via coeficientes.
* **Simplificação de Variáveis:** As variáveis `a`, `b` e `c` saem de cena. A atualização de $x$ passa a ser $x_{novo} = \phi(x)$.
* **Critério de Parada:** O erro é calculado pela diferença entre a iteração atual e a anterior ($|x_{k+1} - x_k|$).

```python
def plotagem(vetA,vetB,titulo):
  plt.arrow(0, 0, vetA[0], vetA[1], color='blue',linewidth=1,head_width=0.08,head_length=0.1,length_includes_head=True,label='Vetor A'), plt.arrow(0, 0, vetB[0], vetB[1], color='red',linewidth=1,head_width=0.08,head_length=0.1,length_includes_head=True,label='Vetor B')
  plt.title(f'{titulo}')
  plt.axis('equal')
  plt.grid(),plt.legend()
  plt.show()
def noventa_graus_rot(vetA,vetB):
  matriz_de_rot = np.array([[0, 1],
                                [-1,  0]])
  plotagem(matriz_de_rot@vetA,matriz_de_rot@vetB,"vetores rotacionados em 90º")
  return (matriz_de_rot@vetA,matriz_de_rot@vetB)
def shear(vetA,vetB,K):
  matriz_de_Shear = np.array([
    [1, K],
    [0, 1]
])
  plotagem(matriz_de_Shear@vetA,matriz_de_Shear@vetB,"shear")
  return (matriz_de_Shear@vetA,matriz_de_Shear@vetB)
def arbitrarios(matrizrot,matrizshear):
  A_x,A_y,B_x,B_y = int(input("\tDigite o valor de A_x (coordenada x do Vetor ARBITRÁRIO A): ")),int(input("\tDigite o valor de A_x (coordenada y do Vetor ARBITRÁRIO A): ")),int(input("\tDigite o valor de B_x (coordenada x do Vetor ARBITRÁRIO B): ")),int(input("\tDigite o valor de B_x (coordenada y do Vetor ARBITRÁRIO B): "))
  vetarbA,vetarbB = np.array([A_x,A_y]),np.array([B_x,B_y])
  plotagem(vetarbA,vetarbB,"vetores arbitrários")
  vetarbA90,vetarbB90 = matrizrot @ vetarbA, matrizrot @ vetarbB
  vetarbAshear,vetarbBshear = matrizshear @ vetarbA,matrizshear @ vetarbB
  print("o vetor multiplicado pela matriz rotacionada é:\t",vetarbA90)
  print(vetarbB90)
  print("o vetor multiplicado pela matriz SHEAR é:\t",vetarbAshear)
  print(vetarbBshear)
  plotagem(vetarbA90,vetarbB90,"vetores arbitrários multiplicados\n pela matriz transformação rotacionada")
  plotagem(vetarbAshear,vetarbBshear,"vetores arbitrários multiplicados\n pela matriz transformação SHEAR")
```

```python
A_x,A_y,B_x,B_y = int(input("\tDigite o valor de A_x (coordenada x do Vetor A): ")),int(input("\tDigite o valor de A_x (coordenada y do Vetor A): ")),int(input("\tDigite o valor de B_x (coordenada x do Vetor B): ")),int(input("\tDigite o valor de B_x (coordenada y do Vetor B): "))
vetor_A,vetor_B = np.array([A_x,A_y]),np.array([B_x,B_y])
plotagem(vetor_A,vetor_B,"vetores sem alterações")
rotvetA, rotvetB = noventa_graus_rot(vetor_A,vetor_B)
shearvA, shearvB = shear(rotvetA, rotvetB,int(input("\n\tinsira o valor da constante k:\t")))
```

```python
matrizrotacionada = np.array([
    rotvetA,
   rotvetB
])
matrizshear = np.array([
    shearvA,
    shearvB
])
print(f"\nmatriz transformação rotacionada :\t\n {matrizrotacionada}")
print(f"\nmatriz transformação SHEAR :\t\n {matrizshear}")
arbitrarios(matrizrotacionada ,matrizshear)
```

### SciPy `linregress`

A função `stats.linregress(x, y)` calcula uma **regressão linear simples**. Ela retorna um objeto contendo os coeficientes que definem a relação linear entre duas variáveis:

1. **Slope (Inclinação):** Taxa de variação de $y$ em relação a $x$.
2. **Intercept (Intercepto):** Onde a reta cruza o eixo $y$ ($x=0$).
3. **R-value:** Coeficiente de correlação de Pearson (proximidade de 1 ou -1 indica ajuste perfeito).
4. **P-value:** Nível de significância (testa a hipótese de que a inclinação é zero).
5. **Standard Error:** Erro padrão do gradiente estimado.

**Aplicação no Código:**
A reta vermelha é gerada pela fórmula:
`y_pred = intercept + slope * x_array`

### scipy.stats.t.interval no Código B

Ni codigo, a função calcula a **margem de erro** ao redor da média para diferentes níveis de probabilidade ($95\%$ a $99\%$).

**O que cada argumento faz no seu `for`:**
1. **`i` (alpha):** É o nível de confiança atual da iteração (ex: 0.95). Ele define a largura da "Área de Confiança" (sombra vermelha).
2. **`df` (Graus de Liberdade):** Definido como `len - 1`. Indica quão robusta é a amostra; quanto maior, mais a distribuição se aproxima de uma Normal.
3. **`loc` (Média):** O ponto central (linha amarela), que é a média das idades geradas.
4. **`scale` (Erro Padrão):** Você usou `stats.sem`, que calcula o erro padrão da média. Ele dita a dispersão do intervalo.

**Comportamento esperado:**
À medida que a `interação` aumenta e o valor de `i` sobe de 0.95 para 0.99, você verá as **linhas vermelhas se afastarem**, pois para ter mais certeza (confiança) de que a média está lá dentro, o intervalo precisa ser maior.

### 1. Hipótese Nula ($H_0$) para o código
A **Hipótese Nula** é o ponto de partida: ela sempre assume que "não há nada de especial acontecendo".
* **$H_0$:** As médias das idades da `populacao_teste` e da `populacao_controle` são **iguais**. Qualquer diferença observada é apenas ruído aleatório.

### 2. O que é o P-Value (Valor-p)?
O **p-value** é a probabilidade de você encontrar esses resultados (ou algo mais extremo) assumindo que a Hipótese Nula é verdadeira.

* **P-Value Pequeno ($\leq 0.05$):** Você rejeita a $H_0$. Conclusão: Há uma diferença estatisticamente significativa entre os grupos.
* **P-Value Grande ($> 0.05$):** Você falha em rejeitar a $H_0$. Conclusão: Não há evidências de que os grupos sejam diferentes.

```python
#questão F

import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

x_array = [1.5,2.3,6,10]
y_array = [0.5,5.4,3,4]

X = np.array(x_array).reshape(-1, 1)
y = np.array(y_array).reshape(-1, 1)

model = LinearRegression()
model.fit(X, y)
b = model.intercept_[0]
a_orig = model.coef_[0][0]


X_with_const = sm.add_constant(X)
results = sm.OLS(y, X_with_const).fit()
print(results.summary())

as_variados = np.linspace(a_orig - 2, a_orig + 2, 6)

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.5, label='Dados Reais')

for a in as_variados:
    y_pred = b + a * X
    plt.plot(X, y_pred, label=f'a = {a:.2f}')

plt.title("Variação da Inclinação (a) com Intercepto (b) Fixo")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

### O que o código faz (Análise de Regressão e Sensibilidade)

O script utiliza duas das bibliotecas mais poderosas para ciência de dados: **Scikit-learn** (para construir o modelo) e **Statsmodels** (para a análise estatística detalhada).

1. **Construção do Modelo (`LinearRegression`)**:

3. **Simulação de Cenários (`for a in as_variados`)**:
   Esta é a parte da "previsão variada". O código fixa o ponto de partida da reta (`b`) e gera 6 novas retas com inclinações diferentes ao redor da ideal.
   * Isso mostra visualmente o **custo do erro**: o quanto a previsão mudaria se a taxa de crescimento fosse levemente diferente da original.

```python
#questão G

import numpy as np
A_exemplo = np.array([[3, -4, 1], [1, 2, 2], [4, 0, -3]])
b_exemplo = np.array([9, 3, -2])

A = A_exemplo.copy().astype(float)
b = b_exemplo.copy().astype(float)
n = len(A)
p = np.arange(n)

print("Matriz inicial A:\n", A)
print("Vetor inicial b:", b)
print("-" * 30)

for k in range(n - 1):
    r = np.argmax(np.abs(A[k:, k])) + k
    pv = A[r, k]

    if pv == 0:
        print("Matriz Singular")

    if r != k:
        print(f"Trocando linha {k} com linha {r} (Pivô: {pv})")
        A[[k, r]] = A[[r, k]]
        p[k], p[r] = p[r], p[k]
        print("A após troca:\n", A)

    for i in range(k + 1, n):
        m = A[i, k] / A[k, k]
        A[i, k] = m
        for j in range(k + 1, n):
            A[i, j] = A[i, j] - m * A[k, j]

    print(f"A após eliminação na coluna {k}:\n", A)

print("-" * 30)
c = np.zeros(n)
for i in range(n):
    c[i] = b[p[i]]
print("Vetor c (b reordenado):", c)

y = np.zeros(n)
for i in range(n):
    soma = 0
    for j in range(i):
        soma += A[i, j] * y[j]
    y[i] = c[i] - soma
print("Vetor y (Ly = c):", y)

x = np.zeros(n)
for i in range(n - 1, -1, -1):
    soma = 0
    for j in range(i + 1, n):
        soma += A[i, j] * x[j]
    x[i] = (y[i] - soma) / A[i, i]
print("Vetor x (Ux = y):", x)


print(x)
```

### Decomposição LU com Pivoteamento Parcial

O objetivo deste script é transformar a matriz original $A$ em duas matrizes triangulares ($L$ e $U$) para encontrar os valores das variáveis desconhecidas no vetor $x$.

1. **Pivoteamento Parcial (`np.argmax`):**
   Antes de cada eliminação, eu busco o maior valor absoluto na coluna atual (`pv`). Eu troco as linhas para garantir que o divisor ($A[k,k]$) seja o maior possível. Isso evita divisões por zero (matriz singular) e aumenta a precisão numérica.

2. **Fatoração na própria matriz:**
   Em vez de criar novas matrizes, eu guardo os multiplicadores $m$ na parte inferior de $A$ (formando a matriz $L$) e os resultados da eliminação na parte superior (formando a matriz $U$).

3. **Resolução em duas etapas:**
   * **Ly = c:** Resolvemos o sistema para um vetor intermediário $y$ usando substituição progressiva.
   * **Ux = y:** Resolvemos o sistema final para $x$ usando substituição regressiva.

**Matriz Singular:**
O código verifica se o pivô é zero. Se for, a matriz é **singular**, o que significa que o sistema não tem solução única (as linhas são linearmente dependentes ou contraditórias).

### Verificação de Convergência

O código analisa a "força" da diagonal principal da matriz em relação aos demais elementos de cada linha.

1. **Cálculo de Peso (`alfa`):** Para cada linha, eu somo os valores absolutos dos elementos que não estão na diagonal e divido pelo valor absoluto do elemento que está na diagonal.
   
2. **Análise do Ponto Crítico (`alfa_max`):** O código busca o maior desses pesos. Se o maior peso for inferior a 1, os cálculos futuros serão estáveis.

3. **Veredito:** * **Sucesso:** O erro vai diminuir a cada iteração até atingir a `tolerancia` definida.
   * **Risco:** Os números podem "explodir" ou nunca chegar a um resultado real, pois a diagonal não é dominante o suficiente.

```python
#questao ultima nao aguento mais AAAAAAAAAAAAAAAAAAAAAA
import numpy as np
import sys
A_exemplo =  np.array([
    [5, 1, 1],
    [3, 4, 1],
    [3, 3, 6]
])
A = A_exemplo.copy().astype(float)
b_exemplo = np.array([5, 6, 0])
B = b_exemplo.copy().astype(float)
dimensao_i, dimensao_j = A_exemplo.shape
n = dimensao_i
beta = [0 for i in range(dimensao_i)]
beta_max = 0
for i in range(n):
      soma_atualizados = 0
      soma_restantes = 0

      for j in range(i):
          soma_atualizados += abs(A[i, j]) * beta[j]

      for j in range(i + 1, n):
            soma_restantes += abs(A[i][j])

      beta[i] = (soma_atualizados + soma_restantes) / abs(A[i, i])

      if beta[i] > beta_max:
          beta_max = beta[i]

print("Betas calculados: ", beta)

if beta_max < 1:
      print("Verdadeiro (Converge) Sassenfeld")
else:
      print("Falso (Não se pode garantir a convergência) Sassenfeld")

#criteiro das linhas
convergencia = True
for i in range(n):
    soma_linha = 0
    valor_diagonal = abs(A[i, i])

    if valor_diagonal == 0:
        convergencia = False
        break

    for j in range(n):
        if i != j:
            soma_linha += abs(A[i, j])

    if soma_linha / valor_diagonal >= 1:
        convergencia = False
        break

if convergencia:
    print("Verdadeiro (Matriz é estritamente diagonal dominante)")
else:
    print("Falso (Critério não satisfeito nesta linha)")
```

### Validação de Convergência: Linhas vs. Sassenfeld

O código executa dois testes de segurança para garantir que, ao resolver o sistema, o erro se aproxime de zero.

#### 1. Critério das Linhas (Mais Simples)
Este é o teste mais rígido. Para cada linha, eu verifico se o elemento da diagonal é maior que a soma de todos os outros elementos daquela linha.
* **Resultado:** Se a soma dividida pela diagonal for sempre $< 1$, a matriz é **Estritamente Diagonal Dominante**. É uma garantia de sucesso, mas muitas matrizes boas falham aqui por serem "quase" dominantes.

#### 2. Critério de Sassenfeld (Mais Inteligente/Poderoso)
Este método é mais sofisticado porque ele é **cumulativo**. No cálculo de cada `beta[i]`, eu utilizo os valores de `beta` que acabei de calcular nas linhas anteriores (`soma_atualizados`).
* **Vantagem:** O Sassenfeld consegue provar que o sistema vai convergir mesmo quando o Critério das Linhas diz que "não". Ele reflete melhor o comportamento do método de **Gauss-Seidel**.
* **Lógica:** Se o `beta_max` final for $< 1$, o sistema é estável e convergirá rapidamente.
